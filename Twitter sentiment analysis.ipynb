```python
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()
lemmatizer.lemmatize("hello", 'a')
```

```python
from nltk.corpus import wordnet

def get_simple_tag(tag):
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN
```

```python
from nltk.corpus import stopwords
from string import punctuation
stops = set(stopwords.words('english'))
stops.update(list(punctuation))
stops
```

```python
from nltk import pos_tag
def clean(words):
    output = []
    for w in words:
        if w.lower() not in stops:
            pos = pos_tag([w])[0][1]
            pos = get_simple_tag(pos)
            lw = lemmatizer.lemmatize(w,pos)
            output.append(lw.lower())
    return output
```

```python
documents = [(clean(doc),category) for doc,category in docs]
```

```python
documents[0]
```

```python
training_docs = documents[:1500]
testing_docs = documents[1500:]
```

```python
all_words = []
for a in training_docs:
    all_words += a[0]
len(all_words)
```

```python
import nltk
freq = nltk.FreqDist(all_words)
most_common = freq.most_common(3000)
most_common
feature_set = [i[0] for i in most_common]
feature_set
```

```python
def get_dict(words):
    ans={}
    words_set = set(words)
    for f in feature_set:
        ans[f] = f in words_set
    return ans
```

```python
get_dict(training_docs[0][0])
```

```python
training_data = [(get_dict(doc),category) for doc,category in training_docs]
testing_data = [(get_dict(doc),category) for doc,category in testing_docs]
```

```python
from nltk import NaiveBayesClassifier
```

```python
clf = NaiveBayesClassifier.train(training_data)
```

```python
nltk.classify.accuracy(clf,testing_data)
```
