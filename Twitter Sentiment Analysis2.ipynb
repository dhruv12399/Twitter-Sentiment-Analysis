```python
tweet = pd.read_csv("C:/Users/Dhruv/Desktop/Machine Learning/Datasets/Twitter/0000000000002747_training_twitter_x_y_train.csv")
tweet_test = pd.read_csv("C:/Users/Dhruv/Desktop/Machine Learning/Datasets/Twitter/0000000000002747_test_twitter_x_test.csv")
Y = np.array(tweet['airline_sentiment'])
X = np.array(tweet['text'])
X_test = np.array(tweet_test['text'])
X_test
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
from nltk.corpus import wordnet

def get_simple_tag(tag):
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN
```

```python
from nltk.corpus import stopwords
from string import punctuation
stops = set(stopwords.words('english'))
stops.update(list(punctuation))
stops
```

```python
from nltk import pos_tag
def clean(words):
    output = []
    for w in words:
        if w.lower() not in stops:
            pos = pos_tag([w])[0][1]
            pos = get_simple_tag(pos)
            lw = lemmatizer.lemmatize(w,pos)
            output.append(lw.lower())
    return output
```

```python
test = []
import nltk
for tw in X:
    test.append(" ".join(clean(nltk.word_tokenize(tw))))
test
```

```python
tw_test = []
for tw in X_test:
    tw_test.append(" ".join(clean(nltk.word_tokenize(tw))))
tw_test
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
cv = TfidfVectorizer(max_features=1700,max_df=0.8, ngram_range=(1,3))
x_train = cv.fit_transform(test)
x_test = cv.transform(tw_test)
x_test.todense()
```
